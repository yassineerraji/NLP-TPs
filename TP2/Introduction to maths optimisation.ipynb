## Exercise 1: Function Plotting and Analytical Determination of Critical Points

Determine _analytically_ the coordinates and nature of the critical points of the following functions:

- $$f_1(x,y) = x^3 + y^3 -3x -12y$$
- $$f_2(x,y) = x^3 -12xy + 8y^3$$
- $$f_3(x,y) = {(x-1)}^2 + 10 {(x^2-y)}^2$$

**_Vocabulary:_** An "analytical" determination means that the calculation is done "by hand", without using numerical tools or formal computation.

```python
# Core scientific Python stack used throughout this TP
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")

# Only needed for a couple of scalar constants (e.g., sqrt(3))
from math import sqrt
```

```python
def f1(x, y):
    """f1(x,y) = x^3 + y^3 - 3x - 12y."""
    return x**3 + y**3 - 3 * x - 12 * y


def f2(x, y):
    """f2(x,y) = x^3 - 12xy + 8y^3."""
    return x**3 - 12 * x * y + 8 * y**3


def f3(x, y):
    """f3(x,y) = (x-1)^2 + 10*(x^2 - y)^2."""
    return (x - 1) ** 2 + 10 * (x**2 - y) ** 2


# --- Small plotting helpers (to avoid repeating the same boilerplate) ---

def make_grid(xmin=-4, xmax=4, nx=41, ymin=-4, ymax=4, ny=41):
    """Return a 2D grid (X, Y) built from 1D linspaces."""
    x1d = np.linspace(xmin, xmax, nx)
    y1d = np.linspace(ymin, ymax, ny)
    return np.meshgrid(x1d, y1d)


def contour_plot(f, xmin=-4, xmax=4, nx=41, ymin=-4, ymax=4, ny=41, n_iso=21):
    """Plot contour lines of a bivariate function on a rectangular domain."""
    X, Y = make_grid(xmin=xmin, xmax=xmax, nx=nx, ymin=ymin, ymax=ymax, ny=ny)
    plt.contour(X, Y, f(X, Y), n_iso)
    plt.title('Contour Lines')
    plt.xlabel('x values')
    plt.ylabel('y values')
    plt.grid()
    plt.axis('square')
```

**3.3** The instruction block below provides the template for the function _df1_, which calculates and returns the gradient of $f_1$. Complete this instruction block so that _df1_ does what is expected. Then use this function to verify that the gradient of $f_1$ is indeed zero at the critical points.

```python
def df1(x, y):
    """Gradient of f1.

    Returns [df/dx, df/dy]. Works with scalars and NumPy arrays.
    """
    df_dx = 3 * x**2 - 3
    df_dy = 3 * y**2 - 12
    return [df_dx, df_dy]
```

```python
# Verify that the gradient is zero at the critical points.
# From df/dx = 3x^2 - 3 and df/dy = 3y^2 - 12:
#   df/dx = 0 -> x = 1
#   df/dy = 0 -> y = 2
# Hence the 4 critical points: (1,2), (1,-2), (-1,2), (-1,-2)

critical_points = [(1, 2), (1, -2), (-1, 2), (-1, -2)]

for pt in critical_points:
    print(f"Gradient at {pt}:", df1(*pt))
```

**3.4** The instruction block below provides the template for the function _d2f1_, which calculates and returns the Hessian matrix of $f_1$. Complete this instruction block so that _d2f1_ does what is expected. Then use this function to verify the nature of the critical points of $f_1$.

```python
def d2f1(x, y):
    """Hessian matrix of f1.

    For f1(x,y) = x^3 + y^3 - 3x - 12y:
      df/dx = 6x, df/dy = 6y, mixed derivatives are 0.

    Returns a 2x2 (nested-list) matrix.
    """
    return [[6 * x, 0], [0, 6 * y]]
```

```python
# Template: eigenvalues of a 2x2 matrix (here, a simple example)
matrice = [[1, 2], [3, 4]]
valProp = np.linalg.eigvals(matrice)

print('Eigenvalues of the matrix {} :'.format(matrice))
print('  vp1 = {}'.format(valProp[0]))
print('  vp2 = {}'.format(valProp[1]))
```

```python
# Determine the nature of a critical point using the Hessian eigenvalues.
# (Example shown for the point (-1, -2).)

# Hessian at the chosen critical point
matrice1 = d2f1(-1, -2)

# Eigenvalues: if both < 0 -> local maximum, both > 0 -> local minimum,
# mixed signs -> saddle point.
valProp = np.linalg.eigvals(matrice1)

print('Eigenvalues of the matrix {} :'.format(matrice1))
print('  vp1 = {}'.format(valProp[0]))
print('  vp2 = {}'.format(valProp[1]))
```
